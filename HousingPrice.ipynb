```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading files
for dirname, _, filenames in os.walk('/Users/hassan/Desktop/Data Sci/KaggleHousingPrices/data'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
# Loading and sorting datasets
train = pd.read_csv('/Users/hassan/Desktop/Data Sci/KaggleHousingPrices/data/train.csv')
test = pd.read_csv('/Users/hassan/Desktop/Data Sci/KaggleHousingPrices/data/test.csv')
test['SalePrice'] = np.NaN
tt_data = pd.concat([train, test],axis=0,sort=False)
```

```python
# Transpose of training data statistics for easier viewability
print("Training data shape: ", train.shape)
train.describe().T
```

```python
print("Testing data shape: ", test.shape)
print("All data shape: ", tt_data.shape)
test.describe().T
```

```python
# Numerical features of dataset
num_data = [c for c in train.columns if train[c].dtype !='O']
cat_data = [w for w in train.columns if train[w].dtype == 'O']

print("Numerical Feature count: ", len(num_data))
num_data
```

```python
print("Categorical Feature count: ", len(cat_data))
cat_data
```

```python
nullCounter = tt_data.isna().sum().sort_values(ascending=False)/len(tt_data)
plt.figure(figsize=(16,10))
plt.title('Missing data bar chart')
plt.xticks(np.arange(len(nullCounter))+0.5, nullCounter.index, rotation='vertical')
plt.ylabel('fraction of rows with null values')
plt.bar(np.arange(len(nullCounter)), nullCounter)
```
